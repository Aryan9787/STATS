```python
import matplotlib.pyplot as plt

# Data for 1983 (Primary sector, up to middle level)
data_1983 = [2.8, 10.0, 8.7, 6.8, 10.5, 6.6, 10.8, 6.9]

# Data for 1993-94 (Primary sector, up to middle level)
data_1993_94 = [4.8, 13.8, 8.7, 3.1, 5.0, 5.1, 8.5, 3.4]

# ------------------------------------------------------------
# Create Box Plot
# ------------------------------------------------------------
plt.figure(figsize=(8, 6))
plt.boxplot([data_1983, data_1993_94],
             tick_labels=["1983", "1993-94"],
             patch_artist=True,
             boxprops=dict(facecolor="gold", color="black"),
             medianprops=dict(color="red"),
             whiskerprops=dict(color="black"),
             capprops=dict(color="black"))

# ------------------------------------------------------------
# Graph Details
# ------------------------------------------------------------
plt.title("Distribution of Workers (Primary Sector, Up to Middle Level Education)")
plt.ylabel("Percentage (%)")
plt.grid(True, linestyle="--", alpha=0.6)
plt.show()
```
