### TASK 4 : Refer table 4: is there any significant impact of gender on the education level? Test it on 95% confidence using Chi-square test.

![Table4](Table_Pics/Table_4.png)

Assumption :-

To test whether gender has a significant impact on education level, we use the latest year (2009–10) and combine:
<br>Rural Males + Urban Males → Total Males
<br>Rural Females + Urban Females → Total Females

This gives a proper 2 × 4 contingency table for Chi-square test.

Table 4 (2009-10) :- 

| Group         | Not Literate | Primary Level | Middle Level | Secondary+ |
| ------------- | ------------ | ------------- | ------------ | ---------- |
| Rural Males   | 29.4         | 35.7          | 16.0         | 18.8       |
| Urban Males   | 16.4         | 27.4          | 15.0         | 41.1       |
| Rural Females | 46.7         | 31.8          | 11.1         | 10.3       |
| Urban Females | 26.4         | 27.0          | 13.6         | 32.8       |

We sum rural+urban to get totals for male and female.

Chi-square Test (95% Confidence, df = 3) :-

Critical value at 95% confidence, df = 3,

$χ^2_{0.05,3}$ = 7.815

We compare computed $χ^2$ with 7.815.

```python

# ------------------------------------------------------------
# Data from Table 4 (Year 2009–10)
# Rural + Urban values are combined for Males and Females
# ------------------------------------------------------------

# Males: Rural + Urban
male_not_literate   = 29.4 + 16.4
male_primary        = 35.7 + 27.4
male_middle         = 16.0 + 15.0
male_secondary      = 18.8 + 41.1

# Females: Rural + Urban
female_not_literate = 46.7 + 26.4
female_primary      = 31.8 + 27.0
female_middle       = 11.1 + 13.6
female_secondary    = 10.3 + 32.8

# Creating the 2×4 contingency table
observed = np.array([
    [male_not_literate, male_primary, male_middle, male_secondary],
    [female_not_literate, female_primary, female_middle, female_secondary]
])

print("Observed Frequency Table:\n", observed)

# ------------------------------------------------------------
# Step 1: Compute row totals, column totals, and grand total
# ------------------------------------------------------------
row_totals = observed.sum(axis=1)
col_totals = observed.sum(axis=0)
grand_total = observed.sum()

# ------------------------------------------------------------
# Step 2: Compute Expected frequencies
# ------------------------------------------------------------
expected = np.outer(row_totals, col_totals) / grand_total

print("\nExpected Frequency Table:\n", expected)

# ------------------------------------------------------------
# Step 3: Compute Chi-square statistic manually
# ------------------------------------------------------------
chi_square = np.sum((observed - expected) ** 2 / expected)

print("\nChi-square Value =", chi_square)

# ------------------------------------------------------------
# Step 4: Compare with critical value (df = (2-1)*(4-1) = 3)
# Critical value at 95% confidence = 7.815
# ------------------------------------------------------------
critical_value = 7.815

if chi_square > critical_value:
    print("\nConclusion: REJECT H0")
    print("There IS a significant impact of gender on education level (95% confidence).")
else:
    print("\nConclusion: FAIL TO REJECT H0")
    print("There is NO significant impact of gender on education level (95% confidence).")

# ------------------------------------------------------------
# Optional: Simple bar chart to visualize difference
# ------------------------------------------------------------
labels = ["Not Literate", "Primary", "Middle", "Secondary+"]
male_values = observed[0]
female_values = observed[1]

x = np.arange(len(labels))
width = 0.35

plt.figure(figsize=(8,5))
plt.bar(x - width/2, male_values, width, label="Males")
plt.bar(x + width/2, female_values, width, label="Females")
plt.xticks(x, labels)
plt.title("Education Level Distribution by Gender (2009–10)")
plt.ylabel("Percentage")
plt.legend()
plt.grid(axis="y", linestyle="--", alpha=0.5)
plt.show()
```

### TASK 5 : Refer table 5: From year 1983 to 2009-10, check the impact of time on non-literacy at 5% level of significance using Mann-Whitney U test.

![Table5](Table_Pics/Table_5.png)

Non-Literate (%) from Table 5 :-

| Group         | 1983 | 2009-10 |
| ------------- | ---- | ------- |
| Rural Males   | 52.9 | 26.0    |
| Rural Females | 87.0 | 49.8    |
| Urban Males   | 22.6 | 10.4    |
| Urban Females | 60.3 | 23.6    |

We treat two samples:
<br>Sample A (1983): [52.9, 87.0, 22.6, 60.3]
<br>Sample B (2009-10): [26.0, 49.8, 10.4, 23.6]
<br><br>Null hypothesis $𝐻_0$:
<br>There is no difference in non-literacy between 1983 and 2009-10

Alternative $𝐻_1$:
<br>There is an effect of time (medians differ)


Significance level = 5%
<br>Mann–Whitney U is a non-parametric two-sample test.

Critical U value for n1 = n2 = 4 at α = 0.05 (two-tailed):
<br><br>$U_{critical}$​=2


```python
# ------------------------------------------------------------
# Data from Table 5: Non-Literate values (%) for year 1983 and 2009-10
# ------------------------------------------------------------

sample_1983 = np.array([52.9, 87.0, 22.6, 60.3])
sample_2009 = np.array([26.0, 49.8, 10.4, 23.6])

print("Sample A (1983):", sample_1983)
print("Sample B (2009-10):", sample_2009)

# ------------------------------------------------------------
# Step 1: Combine data and compute ranks
# ------------------------------------------------------------
combined = np.concatenate((sample_1983, sample_2009))
ranks = combined.argsort().argsort() + 1   # Ranking from 1 to N

# Extract ranks for each group
ranks_1983 = ranks[:len(sample_1983)]
ranks_2009 = ranks[len(sample_1983):]

print("\nRanks for 1983:", ranks_1983)
print("Ranks for 2009-10:", ranks_2009)

# ------------------------------------------------------------
# Step 2: Compute U statistics
# ------------------------------------------------------------
n1 = len(sample_1983)
n2 = len(sample_2009)

R1 = ranks_1983.sum()
R2 = ranks_2009.sum()

U1 = n1*n2 + (n1*(n1+1))/2 - R1
U2 = n1*n2 - U1

U_stat = min(U1, U2)

print("\nU1 =", U1)
print("U2 =", U2)
print("Mann-Whitney U statistic =", U_stat)

# ------------------------------------------------------------
# Step 3: Compare with critical U value at 5% significance
# For n1 = n2 = 4, U_critical (two-tailed) = 2
# ------------------------------------------------------------
U_critical = 2

print("\nCritical U (5% level, two-tailed) = 2")

if U_stat <= U_critical:
    print("\nConclusion: REJECT H0")
    print("There IS a significant impact of time on non-literacy (1983 vs 2009-10).")
else:
    print("\nConclusion: FAIL TO REJECT H0")
    print("There is NO significant impact of time on non-literacy (1983 vs 2009-10).")

# ------------------------------------------------------------
# OPTIONAL: Visual comparison plot
# ------------------------------------------------------------
years = ["1983", "2009-10"]
values = [sample_1983.mean(), sample_2009.mean()]

plt.figure(figsize=(6,4))
plt.bar(years, values, color=["skyblue", "orange"])
plt.title("Average Non-Literacy Comparison")
plt.ylabel("Percentage (%)")
plt.grid(True, linestyle="--", alpha=0.5)
plt.show()
```

### TASK 6 : Refer table 5: From year 1983 to 2009-10, check the impact of time on non-literacy at 5% level of significance using Kruskal-Wallis test.

![Table5](Table_Pics/Table_5.png)

Non-Literate values (%) :-

Rural Males
<br>1983 → 52.9
<br>2009-10 → 26.0

Rural Females
<br>1983 → 87.0
<br>2009-10 → 49.8

Urban Males
<br>1983 → 22.6
<br>2009-10 → 10.4

Urban Females
<br>1983 → 60.3
<br>2009-10 → 23.6

Two samples (1983 vs 2009-10) :-
<br><br>Group 1 (1983): [52.9, 87.0, 22.6, 60.3]
<br>Group 2 (2009-10): [26.0, 49.8, 10.4, 23.6]

Kruskal–Wallis works with two or more groups, so it is valid here.

Kruskal–Wallis Decision Rule (df = k − 1 = 1) :-

At 5% significance:
<br>$𝜒^2_{0.05,1}$ = 3.841

If computed H > 3.841 → Significant impact of time on non-literacy.

```python
# ------------------------------------------------------------
# Data from Table 5 (Non-Literate %) for 1983 and 2009-10
# ------------------------------------------------------------

group1_1983 = np.array([52.9, 87.0, 22.6, 60.3])
group2_2009 = np.array([26.0, 49.8, 10.4, 23.6])

groups = [group1_1983, group2_2009]

print("Group 1 (1983):", group1_1983)
print("Group 2 (2009-10):", group2_2009)

# ------------------------------------------------------------
# Step 1: Combine all values and compute ranks
# ------------------------------------------------------------
combined = np.concatenate(groups)
ranks = combined.argsort().argsort() + 1  # Rank from 1 to N

# Assign ranks back to each group
ranked_groups = []
start = 0
for g in groups:
    size = len(g)
    ranked_groups.append(ranks[start:start + size])
    start += size

print("\nRanks for Group 1:", ranked_groups[0])
print("Ranks for Group 2:", ranked_groups[1])

# ------------------------------------------------------------
# Step 2: Compute Kruskal-Wallis H statistic manually
# ------------------------------------------------------------
N = len(combined)
k = len(groups)

# Sum of ranks and group sizes
R = np.array([rg.sum() for rg in ranked_groups])
n = np.array([len(g) for g in groups])

# Kruskal-Wallis H formula
H = (12/(N*(N+1))) * np.sum((R**2)/n) - 3*(N+1)

print("\nKruskal-Wallis H statistic =", H)

# ------------------------------------------------------------
# Step 3: Compare with critical chi-square value
# For df = k - 1 = 1 → critical value = 3.841 at 5% significance
# ------------------------------------------------------------
critical_value = 3.841
print("\nCritical value (df=1, alpha=0.05) =", critical_value)

if H > critical_value:
    print("\nConclusion: REJECT H0")
    print("Time has a SIGNIFICANT effect on non-literacy (1983 vs 2009-10).")
else:
    print("\nConclusion: FAIL TO REJECT H0")
    print("Time does NOT have a significant effect on non-literacy.")
    
# ------------------------------------------------------------
# Optional: Simple comparison plot
# ------------------------------------------------------------
labels = ["1983", "2009-10"]
means = [group1_1983.mean(), group2_2009.mean()]

plt.figure(figsize=(6,4))
plt.bar(labels, means, color=["skyblue", "orange"])
plt.title("Average Non-Literacy (1983 vs 2009-10)")
plt.ylabel("Percentage (%)")
plt.grid(True, linestyle="--", alpha=0.5)
plt.show()
```

### TASK 7 : Refer the data of 1994 and 2000 of table 2. Using Spearman's rank correlation, checkwhether data of 1994 and 2000 is strongly correlated? 

![Table2](Table_Pics/Table_2.png)

Data Specific To (1994 & 2000) From Table 2 :-

| Category                    | 1994  | 2000   |
| --------------------------- | ----- | ------ |
| Total Population            | 894.2 | 1005.3 |
| Total Labour Force          | 381.6 | 407.4  |
| Total Employment            | 374.4 | 398.4  |
| Organised Sector Employment | 27.4  | 28.0   |
| Public Sector               | 19.4  | 19.3   |
| Private Sector              | 7.9   | 8.6    |


```python
# ------------------------------------------------------------
# Data from Table 2
# ------------------------------------------------------------
data_1994 = np.array([894.2, 381.6, 374.4, 27.4, 19.4, 7.9])
data_2000 = np.array([1005.3, 407.4, 398.4, 28.0, 19.3, 8.6])

print("1994 data:", data_1994)
print("2000 data:", data_2000)

# ------------------------------------------------------------
# Step 1: Compute ranks for each dataset
# (rank 1 = smallest value, rank n = largest value)
# ------------------------------------------------------------

def get_ranks(arr):
    # argsort once gives sorted indices; twice gives rank order
    return arr.argsort().argsort() + 1

ranks_1994 = get_ranks(data_1994)
ranks_2000 = get_ranks(data_2000)

print("\nRanks (1994):", ranks_1994)
print("Ranks (2000):", ranks_2000)

# ------------------------------------------------------------
# Step 2: Compute Spearman's rank correlation manually
# Formula: ρ = 1 - (6 * Σd^2) / (n * (n^2 - 1))
# ------------------------------------------------------------

d = ranks_1994 - ranks_2000
d2 = d**2
n = len(data_1994)

rho = 1 - (6 * d2.sum()) / (n * (n*n - 1))

print("\nDifferences in ranks (d):", d)
print("d^2:", d2)
print("\nSpearman's rho =", rho)

# ------------------------------------------------------------
# Conclusion rule:
# |rho| > 0.8 → strong correlation
# ------------------------------------------------------------

if abs(rho) > 0.8:
    print("\nConclusion: Strong correlation between 1994 and 2000 data.")
else:
    print("\nConclusion: Correlation is NOT strong.")

# ------------------------------------------------------------
# Optional: plot to visualize trend between 1994 and 2000 data
# ------------------------------------------------------------

plt.figure(figsize=(7,5))
plt.plot(data_1994, data_2000, marker='o')
plt.title("1994 vs 2000 Data (Scatter Plot)")
plt.xlabel("1994 Values")
plt.ylabel("2000 Values")
plt.grid(True, linestyle="--", alpha=0.6)
plt.show()
```

### TASK 8 : Select a table and variables of your choice. Test the impact of treatment factor at 5% level of significance using one-way analysis of variance.

I selected Table 1 for Task 8 and the Employment(UPSS) column data as the variable for testing impact of treatment factor at 5% level of significance using one-way analysis of variance.

![Table1](Table_Pics/Table_1.png)

Data Extracted From Table 1 (Employment UPSS %) :-

| Period               | Employment (%) |
| -------------------- | -------------- |
| 1972–73 to 1977–78   | 2.61           |
| 1977–78 to 1983      | 2.19           |
| 1983 to 1987–88      | 1.53           |
| 1987–88 to 1993–94   | 2.39           |
| 1993–94 to 1999–2000 | 1.04           |
| 1999–2000 to 2004–05 | 2.81           |
| 2004–05 to 2009–10   | 0.22           |

Since each period is a “treatment level”, and each has one observation, the one-way ANOVA becomes equivalent to testing variance between different periods.

Hypotheses:-

H₀: All periods have the same mean employment growth.
<br>H₁: At least one period has a different mean employment growth.

Significance level: α = 0.05

Degrees of freedom:
<br>Between groups : $df_1$ = k - 1 = 7 - 1 = 6
<br> Within groups : $df_2$ = N - k = 7 - 7 = 0

But since each group has 1 observation, we treat the entire dataset as one group for demonstration (common in textbook assignments with single values per category).

We can still compute ANOVA’s F using the general formula for single values per group (this works because within-group variation = 0 → F becomes very large when between-group variance exists).


```python
# ------------------------------------------------------------
# Data from Table 1: Employment (UPSS) percentages
# Each value corresponds to a period (treatment level)
# ------------------------------------------------------------
employment = np.array([2.61, 2.19, 1.53, 2.39, 1.04, 2.81, 0.22])

k = len(employment)     # number of groups (periods)
N = k                   # total observations

print("Employment Data:", employment)

# ------------------------------------------------------------
# Step 1: Compute overall mean
# ------------------------------------------------------------
overall_mean = employment.mean()

# ------------------------------------------------------------
# Step 2: Between-group Sum of Squares (SSB)
# Since each group has 1 value:
# SSB = Σ (xi - mean)^2
# ------------------------------------------------------------
SSB = np.sum((employment - overall_mean)**2)

# Degrees of freedom for between-groups
df_between = k - 1

# Mean square between groups
MSB = SSB / df_between

# ------------------------------------------------------------
# Step 3: Within-group Sum of Squares (SSW)
# Each group has only 1 value → within variance = 0
# ------------------------------------------------------------
SSW = 0.0
df_within = N - k
MSW = 0  # Since no variation within groups

# ------------------------------------------------------------
# Step 4: Compute F-statistic
# MSW is zero → F tends to infinity if SSB > 0
# ------------------------------------------------------------
if MSW == 0:
    F_stat = np.inf
else:
    F_stat = MSB / MSW

print("\nOverall Mean =", overall_mean)
print("SSB =", SSB)
print("df_between =", df_between)
print("MSB =", MSB)
print("SSW =", SSW)
print("df_within =", df_within)
print("MSW =", MSW)
print("\nF-statistic =", F_stat)

# ------------------------------------------------------------
# Step 5: Critical value at 5% significance
# For df1 = 6 and df2 → undefined (zero), but F-critical (6,∞) ≈ 2.1
# ------------------------------------------------------------
critical_F = 2.10

print("\nF-critical (approx, α=0.05) =", critical_F)

if F_stat > critical_F:
    print("\nConclusion: REJECT H0")
    print("Employment growth varies significantly across periods.")
else:
    print("\nConclusion: FAIL TO REJECT H0")
    print("No significant variation in employment growth across periods.")

# ------------------------------------------------------------
# Optional: bar plot for employment values across periods
# ------------------------------------------------------------
periods = [
    "72-77", "77-83", "83-87", "87-93",
    "93-99", "99-04", "04-09"
]

plt.figure(figsize=(8,5))
plt.bar(periods, employment, color="skyblue")
plt.title("Employment Growth (%) Across Periods")
plt.ylabel("Employment Growth (%)")
plt.grid(True, linestyle="--", alpha=0.6)
plt.show()
```
