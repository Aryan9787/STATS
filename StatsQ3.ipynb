```python
import matplotlib.pyplot as plt
import numpy as np

years = [1994, 2000, 2005, 2010]
population = [894.2, 1005.3, 1094.0, 1187.1]

# Ogive (Cumulative Frequency Curve)
cumulative_population = np.cumsum(population)  # cumulative sum for illustration

plt.figure(figsize=(8, 5))
plt.plot(years, cumulative_population, marker='o', color='blue', linestyle='-')
plt.title("Ogive (Cumulative Curve) of Total Population")
plt.xlabel("Year")
plt.ylabel("Cumulative Population (in millions)")
plt.grid(True, linestyle="--", alpha=0.6)
plt.show()

# Pie Chart  Share of each year's population
plt.figure(figsize=(6, 6))
plt.pie(population, labels=years, autopct='%1.1f%%', startangle=90, shadow=True)
plt.title("Distribution of Total Population (19942010)")
plt.show()

```
