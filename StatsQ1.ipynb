```python
import math
import statistics

# Data: Annual population growth rate (%)
population_growth = [2.18, 2.21, 2.13, 2.06, 1.97, 1.71, 1.65]


# 1. Arithmetic Mean
arithmetic_mean = sum(population_growth) / len(population_growth)


# 2. Geometric Mean
product = 1
for value in population_growth:
    product *= value
geometric_mean = product ** (1 / len(population_growth))

# 3. Median
median_value = statistics.median(population_growth)

# 4. Standard Deviation
std_dev = statistics.stdev(population_growth)

# Display Results
print("STATISTICAL ANALYSIS OF POPULATION GROWTH RATE (in %)")
print("-" * 60)
print(f"Data: {population_growth}")
print(f"\nArithmetic Mean     : {arithmetic_mean:.3f}")
print(f"Geometric Mean      : {geometric_mean:.3f}")
print(f"Median              : {median_value:.3f}")
print(f"Standard Deviation  : {std_dev:.3f}")
```
